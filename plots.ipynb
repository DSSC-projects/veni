```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
with open('../v5.pk', 'rb') as f:
    v5 = pickle.load(f)
with open('../v10.pk', 'rb') as f:
    v10 = pickle.load(f)
with open('../v20.pk', 'rb') as f:
    v20 = pickle.load(f)
with open('../v40.pk', 'rb') as f:
    v40 = pickle.load(f)
```

```python
n = 60000 // 64
plt.figure(figsize=(10,7))

labels = ['5','10','20','40']
trainLosses = []

for v in [v5, v10, v20, v40]:
    arr = [[l[0]*n + l[1], l[2]] for l in v[0]]
    trainLosses.append(np.array(arr))

for t, l in zip(trainLosses, labels):
    plt.plot(t[:,0],t[:,1],'.-', label = l)

plt.title('LOGISTIC REGR. - Comparision between different number of directions chosen \n Training')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.grid()
plt.legend()
plt.show()

```
